```python
from google.colab import files
import os
from PIL import Image
import matplotlib.pyplot as plt
from IPython.display import display

def upload_folder():
    print("Please select multiple image files to upload...")
    uploaded_files = files.upload()

    folder_name = "uploaded_images"
    os.makedirs(folder_name, exist_ok=True)

    for filename in uploaded_files.keys():
        file_path = os.path.join(folder_name, filename)
        with open(file_path, "wb") as f:
            f.write(uploaded_files[filename])

    print(f"Images uploaded to: {folder_name}")
    return folder_name

def load_images_from_folder(folder_path, image_format=("png", "jpg", "jpeg")):
    images = []
    filenames = sorted(os.listdir(folder_path))  # Sort for consistency

    for file in filenames:
        if file.lower().endswith(image_format):
            try:
                img_path = os.path.join(folder_path, file)
                img = Image.open(img_path)
                images.append(img)
            except Exception as e:
                print(f"Error loading image {file}: {e}")

    print(f"Loaded {len(images)} images from {folder_path}")
    return images

def display_images(images, num=5):
    if not images:
        print("No images to display.")
        return

    num = min(num, len(images))  # Ensure we don't try to display more images than available
    fig, axes = plt.subplots(1, num, figsize=(15, 5))

    for i in range(num):
        axes[i].imshow(images[i])
        axes[i].axis('off')  # Hide axes for a cleaner view

    plt.show()

def create_gif(images, output_path="output.gif", duration=500):
    if not images:
        print("No images provided for GIF creation.")
        return

    # If only one image, repeat it to create a valid GIF
    if len(images) == 1:
        images = images * 10  # Repeat the image 10 times (or any number you prefer)

    images[0].save(output_path, save_all=True, append_images=images[1:], duration=duration, loop=0)
    print(f"GIF saved as {output_path}")
    display(Image.open(output_path))


# Main Flow
folder_path = upload_folder()

if folder_path:
    images = load_images_from_folder(folder_path)
    display_images(images, num=5)  # Display first 5 images
    create_gif(images, "generated.gif", duration=500)
```
